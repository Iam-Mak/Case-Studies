## Case Study
You are in the Empire State Building , your next move depends on the number of eyes you throw with the dice.<br>
The sample code assumes that you're currently at **step 50** and  you have **100 chances** and you have to reach **step 60**.
You can perfectly code this with an `if-elif-else` construct! <br>
- take 1 step backward if dice shows 1 or 2.
- take 1 step forward if dice shows 3, 4, or 5.
- throw dice again if dice shows 6 and then take that number of steps.
- Can't go below step 0
- **0.1%** chance of falling down the stairs.
- **Bet**: you'll reach step 60


  
 
Can you fill in the missing pieces to finish the script? Lets see..
 
 

In this exercise, we will be using two functions from `numpy` package:
- `seed()`: sets the random seed, so that your results are reproducible between simulations. As an argument, it takes an integer of your choosing. If you call the function, no output will be generated.

- `rand()`: if you don't specify any arguments, it generates a random float between zero and one.

```python
# Import numpy as np
import numpy as np

# Set the seed
np.random.seed(123)

# Generate and print random float
print(np.random.rand())
```

### Roll the dice

```python
# Use randint() to generate integer to simulate a dice
np.random.seed(123)
print(np.random.randint(1,7))
```

### Let's Start -  Determine your next move

```python
# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <=5 :
    step =step +1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

### The Random Walk
It's time to put this code inside a `for loop` so that we can simulate a random walk.

```python
np.random.seed(123)

# Initialize random_walk
random_walk = [0]

# 
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

**However, there's something we haven't thought about - you can't go below 0.** <br>
A typical way to solve problems like this is by using `max()`.

```python
np.random.seed(123)

# Initialize random_walk
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # Use max to make sure step can't go below 0
        step = max(0, step -1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
```

### Visualize the walk

```python
import matplotlib.pyplot as plt
```

```python
# Plot random_walk
plt.plot(random_walk)

plt.ylabel("Steps taken",fontsize=18)
plt.xlabel("100 chance",fontsize=18)
plt.title("Random Walk", loc = 'left',fontsize=20)

# Show the plot
plt.show()
```

### Distribution - Simulate multiple walks
To get an idea about how big your chances are of reaching 60 steps, you can repeatedly simulate the random walk and collect the results.<br>
Let's Simulate random walk 10 times

```python
np.random.seed(123)

# Initialize all_walks
all_walks = []

# Simulate random walk 10 times
for i in range(10) :

    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

## Visualize all walks
`all_walks` is a **list of lists:** every **sub-list** represents a **single random walk**. If you convert this list of lists to a **Numpy array**, you can start making interesting **plots!**

```python
# Convert all_walks to Numpy array: all_walks_array
all_walks_array = np.array(all_walks)

plt.ylabel("Steps taken",fontsize=18)
plt.xlabel("10 random walks",fontsize=18)
plt.title("All Walks", loc = 'left',fontsize=20)


# Plot np_aw 
plt.plot(all_walks_array)
plt.show()

# Clear the figure
plt.clf()
```

Above graph is not usfull as it just joins all the points in 10 walks_array.

Transpose `all_walks_array` by calling `np.transpose()` on `all_walks_array`. Call the result `all_walks_array_trans`. Now every row in np_all_walks represents the position after 1 throw for the 10 random walks.

```python
# Clear the figure
plt.clf()

# Transpose all_walks_array: all_walks_array_trans
all_walks_array_trans = np.transpose(all_walks_array)

plt.ylabel("Steps taken",fontsize=18)
plt.xlabel("100 chance",fontsize=18)
plt.title("10 Random Walks", loc = 'left',fontsize=20)

# Plot all_walks_array_trans and show
plt.plot(all_walks_array_trans)
plt.show()
```

## Implement clumsiness
There's still something we forgot!  You're a bit clumsy and you have a `0.1%` chance of falling down. That calls for another random number generation. Basically, you can generate a random float between 0 and 1. If this value is less than or equal to `0.001`, you should reset step to 0.


```python
np.random.seed(123)

# Simulate random walk 250 times
all_walks = [] 
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot all_walks_array_trans

plt.ylabel("Steps taken",fontsize=18)
plt.xlabel("100 chance",fontsize=18)
plt.title("250 Random Walks", loc = 'left',fontsize=20)


all_walks_array_trans = np.transpose(np.array(all_walks))
plt.plot(all_walks_array_trans)
plt.show()
```

## Plot the distribution
#### What are the odds that you'll reach 60 steps high on the Empire State Building?
Basically, you want to know about the end points of all the random walks you've simulated. These end points have a certain distribution that you can visualize with a histogram.

```python
np.random.seed(123)
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot all_walks_array_trans
all_walks_array_trans = np.transpose(np.array(all_walks))

# Select last row from all_walks_array_trans: ends
ends = all_walks_array_trans[-1,:]

# Plot histogram of ends, display plot

plt.xlabel("Final Step in Random Walk",fontsize=14)
plt.ylabel("Frequency of Final Step",fontsize=14)
plt.title("500 Random Walks", loc = 'left',fontsize=20)


plt.hist(ends,alpha = 0.7)
plt.grid (True)
plt.show()
```

### Calculate the odds

The above histogram was created from a **Numpy array ends**, that contains **500 integers**. Each integer represents the end point of a random walk. To calculate ``the chance that this end point is greater than or equal to 60``, you can count ``the number of integers in ends that are greater than or equal to 60 and divide that number by 500, the total number of simulations.``<br>
``np.mean(ends >= 60)``

```python
np.mean(ends >=60)
```

```python
plt.xlabel("Final Step in Random Walk",fontsize=14)
plt.ylabel("Frequency of Final Step",fontsize=14)
plt.title("500 Random Walks", loc = 'left',fontsize=20)

plt.axvline(x=60,color='red')

plt.hist(ends,alpha = 0.75,facecolor='g')
plt.grid (True)
plt.show()
```
